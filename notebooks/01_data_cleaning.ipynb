```python
import pandas as pd

df_check = pd.read_csv('../data/superstore_clean.csv', parse_dates=['Order Date', 'Ship Date'])

print('Region value counts:')
print(df_check['Region'].value_counts())
print(f'\nSales max:        ${df_check["Sales"].max():,.2f}')
print(f'Profit dtype:     {df_check["Profit"].dtype}')
print(f'Order Date dtype: {df_check["Order Date"].dtype}')
print(f'Quantity min:     {df_check["Quantity"].min()}')
print(f'Total rows:       {len(df_check)}')

assert set(df_check['Region'].unique()) == {'West', 'East', 'South', 'Central'}
assert df_check['Sales'].max() <= 50000
assert (df_check['Quantity'] >= 0).all()
assert str(df_check['Profit'].dtype) == 'float64'
assert 'datetime' in str(df_check['Order Date'].dtype)
print('\nAll validation checks passed.')
```
